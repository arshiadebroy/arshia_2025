Correct Answer: D, 2 raised to the ninety sixth power times as many addresses are available.
-  With 32-bit addressing, IPv4 has 2 raised to the thirty second power possible addresses. With 128-bit addressing, IPv6 has 2 raised to the one hundred twenty eighth power possible addresses.
