## Notes on Cybersecurity 
- Exploring the World of Cybersecurity: Panel Takeaways
What is Cybersecurity?
Cybersecurity is the practice of defending digital systems, networks, and sensitive information from unauthorized access, attacks, and damage. With the increasing reliance on technology, the field has expanded rapidly, becoming essential across nearly every industry.

Meet the Experts
David Tomasz
An FBI agent focused on cybercrime, Tomasz works closely with international agencies to track and apprehend hackers. His work involves identifying and stopping high-level cyber threats, often involving global politics.

Tricia Mercer
Serving on the board of San Diego’s Cyber Center of Excellence, Mercer brings over two decades of cybersecurity experience. Her background combines technical knowledge with strategic business planning, especially in the context of data protection.

Victor Nzeata
A veteran in military cyber operations, Nzeata now trains others in cybersecurity. He transitioned from software development to cyber defense, using his technical skills to secure systems and educate the next generation of professionals.

Major Discussion Points
Current Threat Landscape
David Tomasz spoke about serious cyber threats from foreign nations, like China, especially targeting biotech firms and high-value research.

Victor Nzeata highlighted the risks AI presents, including the potential for misuse in various industries.

Tricia Mercer emphasized phishing attacks, which often exploit individuals’ trust and can lead to major data breaches.

Insider Threats & Real Cases
Tomasz has tracked down notorious hackers involved in massive data theft.

Nzeata discovered crucial information by analyzing leftover code from military operations, helping prevent serious breaches.

Mercer shared a case where a former employee with financial issues continued accessing company databases after taking a leave—illustrating how internal risks can be just as dangerous as external ones.

Zero-Day Vulnerabilities
These are critical flaws in software that have no immediate fix or patch, giving hackers a window to exploit systems undetected. They’re particularly dangerous because they’re unknown until it's too late.

Ethical Issues in Cybersecurity
Ethics play a huge role in cybersecurity. While malicious hackers aim to harm or exploit, ethical hackers (or “white hats”) test systems to improve their defenses. Professionals in the field must be trusted to use their knowledge responsibly, emphasizing the importance of integrity.

Careers in Cybersecurity
What Inspired Their Journeys?
Mercer was captivated by a book on espionage, sparking her interest in cyber defense.

Tomasz didn’t start off liking computers, but after some encouragement from a friend, he discovered his aptitude and passion for the field.

Nzeata transitioned from software development after realizing how impactful cybersecurity could be.

Advice for Students
Build a strong foundation in IT and computer systems.

Network with people already in the field—relationships matter.

Learn by doing—real-world experience is key.

Be dependable, adaptable, and curious.

Biggest Challenges They Faced
Mercer found it tough to secure funding and simplify tech jargon for investors unfamiliar with cybersecurity.

Tomasz said internal betrayal from colleagues can be a major issue.

Nzeata faced personal hardship, being stationed far from home in dangerous environments during military service.

Looking Ahead: The Future of Cybersecurity
Nzeata sees AI becoming even more integrated into cybersecurity, especially in sectors like healthcare. He also mentioned that the field is becoming more accessible in terms of training and costs.

Tomasz warned that AI could lead to job displacement, as automation reduces the need for certain roles.

Mercer believes that more companies will prioritize cyber defense, especially since many developers still don’t know how to secure their code properly.

Final Q&A Highlights
Professionals often use the same tools as hackers but with ethical intent. Tools like NGINX are used for network analysis and “sniffing” vulnerabilities.

Recommended getting certified—especially through free programs like the (ISC)² Certified in Cybersecurity (CC).

Emphasized the value of teamwork—group projects mimic real work environments.

Not all cybersecurity jobs require deep technical skills; many involve complex information management and critical thinking.

Personal Reflection
This panel opened my eyes to how diverse and impactful the field of cybersecurity really is. It’s not just about tech—it’s about human behavior, teamwork, and staying vigilant in a constantly changing world. I was especially inspired by how different each speaker’s path was. Whether through military service, business leadership, or unexpected curiosity, they all found purpose in protecting others. The key lessons I’m walking away with? Stay ethical, keep learning, and don’t underestimate the power of collaboration.
