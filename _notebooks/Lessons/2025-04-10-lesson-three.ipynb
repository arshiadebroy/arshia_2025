## Binary Search Algorithm
- Binary Search is an efficient algorithm used to find the position of a target value within a sorted list. It works by repeatedly dividing the search interval in half.
- Binary Search won't work correctly on unsorted lists.
- Always check the list is sorted before applying the algorithm.
- Start with the entire list.

Find the middle element.

Compare the middle element to the target:

If it matches, return the index.

If the target is less, search the left half.

If the target is greater, search the right half.

Repeat steps 2–3 until the target is found or the range is empty.

FRQ Response:
Why is binary search more efficient for large data than linear search?

Binary search is much more efficient because it repeatedly cuts large data sets in half. We can also explain this using the dictionary example. It is a lot easier to split the book in half and then go either left or right from that point than to read through every single word. Reading through every single word would be an example of linear searching, which is not time or cost-efficient.
What happens if the list isn’t sorted and you try to use binary search?

If the list is not sorted, binary searching would not work, and as a result, the algorithm would most likely return the false target or just completely fail to find the target. In order for binary to be able to divide the data sets provided, the data needs to flow in some sort of logical sense (ascending, descending, etc).
Could you use binary search in a video game leaderboard or streaming service search bar? Why or why not?

Yes, you could use binary search for a video game leaderboard. Video game sites tend to have many users that you can view, and leaderboard data is always sorted. This sorting can be by who has the highest rank, most points, most games played, etc. Because of this, video games have both sorted and large amounts of data, causing binary searching to be applicable. However, it may not be the best choice for a streaming service search bar since the data is a lot more unsorted and randomized based on user preference, trendingness, watch history, etc.
